```python
def distanceCalculator(m, c, x1, y1):
    A = m
    B = -1
    C = c
    return (abs(A * x1 + B * y1 + C)) / (math.sqrt(A * A + B * B))

if __name__ == "__main__":

  import numpy as np
  import math
  from pylab import *

  global BestFit, mAns, bAans
  sum, mAns, bAans = 0,0,0  #declaring the variable integer

    #the random points on graph
  x = np.array([0, 1, 2, 5, 9, 5, 6, 7, 8, 9])
  y = np.array([1, 3, 2, 5, 4, 8, 8, 9, 12, 12])

  BestFit = 999999  # random max number

  sumlist=[]
  # generate m,b value
  for m in np.linspace(5, -5, 100):
      for b in np.linspace(-5, 5, 100):
          for i in range(len(x)):
              sum += distanceCalculator(m, b, x[i], y[i])
          if (sum <= BestFit):
              BestFit = sum
              mAns, bAans = m, b
          sumlist.append(sum)
          sum=0
print('the ans my code shows',mAns,bAans)
plot(x, y, 'yo', x, mAns*x+bAans, '--k')
show()
print(sumlist)
```
