```
import random
import math 
import numpy as np
from matplotlib import pyplot as plt
#generating 100 random number ranging 0 to 10 
x = [random.randrange(1, 10) for i in range(99)]

y = [random.randrange(1, 10) for i in range(99)]
```

```
#check the values inside 100random 2d data
print ("Random number list of x is : " +  str(x))
print ("Random number list of y is : " +  str(y))
plt.scatter(x,y)
```

```
#manually assigning randomly selecting centroids when k=3
m1=[5,1]
m2=[1,9]
m3=[9,1]
```

```
#initializing the individual cluster lists
c1=[[5,1]]
c2=[[1,9]]
c3=[[9,1]]
```

```
#newM calculates the value of new centroid 
#based on the cluster list created in rearrangingClusters()
def newM(c1):
    c1Avg=0
    c1Avg=0
    for i in range(0,len(c1)-1):
            c1Avg+=c1[i][0]
            c1Avg+=c1[i][1]
    aX=int(c1Avg/len(c1))
    print("length of c1 list"+str(len(c1)))
    aY=int(c1Avg/len(c1))
    m1=[aX,aY]   
    print("length of c1: "+str((len(c1))))
    print("new m1 value: "+str(m1)+"\n")    
    return m1

def rearrangingClusters():
         
      for i in range(0,99):
          a=int(x[i])
          b=int(y[i])
          m1Distance=math.sqrt(((m1[0]-a)**2)+((m1[1]-b)**2))
          m2Distance=math.sqrt(((m2[0]-a)**2)+((m2[1]-b)**2))
          m3Distance=math.sqrt(((m3[0]-a)**2)+((m3[1]-b)**2))
          theMinNumber=min(m1Distance,m2Distance,m3Distance)
          print("the min number is:"+ str(theMinNumber))
          if m1Distance==theMinNumber:
              c1.append([a,b])
              print("added "+str([a,b])+"in c1\n")
          elif m2Distance==theMinNumber:
              c2.append([a,b])        
              print("added "+str([a,b])+"in c2\n")
          else:
             c3.append([a,b])
             print("added "+str([a,b])+" in c3\n")
    
if newM(c1)!=m1 and newM(c2)!=m2 and newM(c2)!=m2 :
  m1=newM(c1)            
  m2=newM(c2)
  m3=newM(c3)
  #clearing all the cluster list 
  #and then repopulating the cluster list based on new centroid m1,m2,m3 
  c1.clear()
  c2.clear()
  c3.clear()
  rearrangingClusters()
else:
  m1=newM(c1)            
  m2=newM(c2)
  m3=newM(c3)
```

```
#final centroids     
print("centroid of cluster 1: "+str(m1))
print("centroid of cluster 2: "+str(m2))
print("centroid of cluster 3: "+str(m3))
```
