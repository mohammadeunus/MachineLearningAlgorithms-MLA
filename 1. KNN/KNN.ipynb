### Problem Statement:
Create the K nearest neighbour model that can classify the different species of the Iris flower.


#### Solving steps
- Create the dataset.
- Segment the dataset into train and test set
- Build and fit the model
- Make predictions.

#### Iris Flower Dataset:
The Iris flower dataset is publicly available and $sklearn$ comes with the inbuilt datasets for the iris classification problem.

The data set consists of:
- 150 samples
- 3 labels: species of Iris (Iris setosa, Iris virginica and Iris versicolor)
- 4 features: Sepal length, Sepal width, Petal length, Petal Width in cm

from sklearn import  datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

```python
# import libraries

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# load dataset and split the features and class lables

iris = datasets.load_iris()

x=iris.data
y=iris.target

print("x: \n", x[95:105])
print("y\n", y[95:105])
print(iris)
```

```python
# WE need three parts in a dataset for training and testing
# train set
# dev set
# test set

# Split the dataset into training sets and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4)

print(x.shape)

print(x_train.shape)
print(y_train.shape)

print(x_train)
print(y_train)
```

```python
# Build a knn classifier

knn = KNeighborsClassifier(n_neighbors=21)
knn.fit(x_train, y_train)   # memorizing the values in x and y
```

```python
predictions = knn.predict(x_test)

print(y_test[0:30])
print(predictions[0:30])
```

```python
print(accuracy_score(y_test,predictions))
```
